```python
df = df.drop(columns=['Unnamed: 22', 'fulfilled-by'])
```

```python
df = df.dropna(
    subset=[
        'ship-city',
        'ship-state',
        'ship-postal-code',
        'ship-country'
    ]
)
```

```python
df['currency'] = df['currency'].fillna('INR')
```

```python
df['Promoted'] = df['promotion-ids'].notna().astype(int)
```

```python
def promotion_type(x):
    if pd.isna(x):
        return 'No Promotion'
    elif 'Free Shipping' in str(x):
        return 'Free Shipping'
    elif 'PLCC' in str(x):
        return 'Amazon Financing'
    elif 'Coupon' in str(x):
        return 'Coupon'
    else:
        return 'Other Promotion'

df['Promotion_Type'] = df['promotion-ids'].apply(promotion_type)
```
